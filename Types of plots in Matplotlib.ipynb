```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'sugar': 5, 'flour': 15, 'tea': 20, 'coffee': 10}
bars = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(10, 3), sharey=True)
axs[0].bar(bars, values)
axs[1].scatter(bars, values)
axs[2].plot(bars, values)
fig.suptitle('Categorical Plotting')
```

```python
#Random state for reproducibility
np.random.seed(19680801)

dt= 0.01
t= np.arange(0, 30, dt)

wn1= np.random.randn(len(t))    # white noise 1
wn2= np.random.randn(len(t))    # white noise 2            

# Two signals with a coherent part at 20Hz and a random part
sig1= np.sin(2* np.pi* 20* t)+ wn1
sig2= np.sin(2* np.pi* 20* t)+ wn2

fig, axs= plt.subplots(2, 1)
axs[0].plot(t, sig1, t, sig2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Time')
axs[0].set_ylabel('sig1 and sig2')
axs[0].grid(True)

cxy, f= axs[1].cohere(sig1, sig2, 256, 1. /dt)
axs[1].set_ylabel('Coherence')

fig.tight_layout()
plt.show()
```

```python
N= 20
x= np.linspace(0, 10, 11)
y= [4, 4.7, 10.5, 9.3, 11.8, 12.7, 14.1,  9.9, 17.9, 14.2, 11.1]

#Linear curve, an estimate its y-values and their error.
a, b= np.polyfit(x, y, deg=1)
y_est= a* x+ b
y_err= x.std()* np.sqrt(1/len(x) +
                 (x - x.mean())**2 / np.sum((x - x.mean())**2))

fig, ax= plt.subplots()
ax.plot(x, y_est, '-')
ax.fill_between(x, y_est- y_err, y_est+ y_err, alpha=0.2)
ax.plot(x, y, 'o', color='tab:red')
```
